# Data Visualisation

### Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

### Import file

```python
file_1 = pd.read_csv('Datasets/File_1.csv', low_memory=False)
file_1 = file_1.drop('Unnamed: 0', axis=1)
file_1
```

### Information

```python
file_1.info()
```

```python
file_1.describe()
```

```python
file_1.shape
```

### Transform data

```python
# Convert 'timestamp' and 'last_position' to datetime
file_1['timestamp'] = pd.to_datetime(file_1['timestamp'])
file_1['last_position'] = pd.to_datetime(file_1['last_position'])
file_1
```

```python
file_1 = file_1[(file_1['longitude'] >= -10) & (file_1['longitude']<= 10)]
file_1
```

### Create scatterplot

```python
plt.scatter(file_1['longitude'], file_1['latitude'], c=file_1['baro_altitude'], cmap='viridis', alpha=0.5)
plt.title('Flight Trajectories')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.colorbar(label='Baro altitude')
plt.show()
```

### Create boxplot of altitude

```python
plt.figure(figsize=(10, 6))
plt.boxplot(file_1['altitude'].dropna())
plt.title('Boxplot van Altitude')
plt.xlabel('Altitude (feet)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.boxplot(file_1['baro_altitude'].dropna())
plt.title('Boxplot van Baro altitude')
plt.xlabel('Baro altitude (feet)')
plt.show()
```

```python
print(file_1['baro_altitude'].max())
print(file_1['baro_altitude'].min())
```

### Pie Chart of On-Ground Status

```python
onground_counts = file_1['onground'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(onground_counts, labels=onground_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('On-Ground Status')
plt.show()
```

### Create boxplot of ground speed

```python
plt.figure(figsize=(8, 6))
plt.boxplot(file_1['ground_speed'].dropna())
plt.title('Boxplot of Ground Speed')
plt.xlabel('Ground Speed')
plt.show()
```

```python
print(file_1['ground_speed'].max())
print(file_1['ground_speed'].min())
```

### Histogram of altitude

```python
plt.figure(figsize=(10, 6))
plt.hist(file_1['altitude'].dropna(), bins=50, color='blue', alpha=0.7)
plt.title('Histogram of Altitude')
plt.xlabel('Altitude (feet)')
plt.ylabel('Frequency')
plt.show()
```

### Time series of baro altitude

```python
file_1['timestamp'] = pd.to_datetime(file_1['timestamp'])
plt.figure(figsize=(12, 6))
plt.plot(file_1['timestamp'], file_1['baro_altitude'])
plt.title('Baro altitude Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Baro altitude (feet)')
plt.show()
```

### Scatterplot of ground speed vs. baro altitude

```python
plt.figure(figsize=(10, 8))
plt.scatter(file_1['ground_speed'], file_1['baro_altitude'], alpha=0.5)
plt.title('Scatter Plot of Ground Speed vs. Altitude')
plt.xlabel('Ground Speed')
plt.ylabel('Altitude (feet)')
plt.show()
```

### Boxplot of vertical rate

```python
plt.figure(figsize=(10, 6))
plt.boxplot(file_1['vertical_rate'].dropna())
plt.title('Boxplot of vertical rate')
plt.xlabel('Vertical rate (meters/second)')
plt.show()
```

```python
print(file_1['vertical_rate'].max())
print(file_1['vertical_rate'].min())
```

### Violinplot of vertical rate

```python
drempel = 10000

gefilterde_waarden = file_1['vertical_rate'].dropna()
gefilterde_waarden = gefilterde_waarden[(gefilterde_waarden >= -drempel) & (gefilterde_waarden <= drempel)]
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(y=gefilterde_waarden)
plt.title('Violinplot of vertical rate')
plt.ylabel('Vertical rate (meters/second)')

# Stel de y-limiet in om waarden boven de drempel te voorkomen
plt.ylim(-drempel, drempel)

plt.show()
```
