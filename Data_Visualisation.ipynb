# Data Visualisation

### Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

### Import file

```python
# Load data from a CSV file into a Pandas DataFrame
file_1 = pd.read_csv('Datasets/Flights/File_1.csv', low_memory=False)
# Drop the 'Unnamed: 0' column that is an unwanted index column
file_1 = file_1.drop('Unnamed: 0', axis=1)
# Display the contents of the imported data
file_1.head()
```

### Information

```python
# Print information about the DataFrame, including the data types and number of non-null values
file_1.info()
```

```python
# Generate descriptive statistics, which include mean, std, min, and max values for numerical columns
file_1.describe()
```

```python
# Print the shape (number of rows and columns) of the DataFrame
file_1.shape
```

### Transform data

```python
# Convert the 'timestamp' and 'last_position' columns from string to datetime objects for easier handling of dates
file_1['timestamp'] = pd.to_datetime(file_1['timestamp'])
file_1['last_position'] = pd.to_datetime(file_1['last_position'])
file_1
```

```python
# Filter the data to include only longitudes between -20 and 20 degrees
file_1 = file_1[(file_1['longitude'] >= -20) & (file_1['longitude']<= 20)]
file_1
```

### Create scatterplot

```python
# Create a scatter plot of flights' positions with color indicating barometric altitude
plt.scatter(file_1['longitude'], file_1['latitude'], c=file_1['baro_altitude'], cmap='viridis', alpha=0.5)
plt.title('Flight Trajectories')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
# Add a color bar to show the altitude scale
plt.colorbar(label='Baro altitude')
plt.show()
```

### Create boxplot of altitude

```python
# Create a boxplot for the 'altitude' column to show distribution, including median, quartiles, and outliers
plt.figure(figsize=(10, 6))
plt.boxplot(file_1['altitude'].dropna())
plt.title('Boxplot of Altitude')
plt.xlabel('Altitude (feet)')
plt.show()
```

```python
# Repeat the above for 'baro_altitude'
plt.figure(figsize=(10, 6))
plt.boxplot(file_1['baro_altitude'].dropna())
plt.title('Boxplot van Baro altitude')
plt.xlabel('Baro altitude (feet)')
plt.show()
```

```python
# Print the maximum and minimum values of barometric altitude for reference
print(file_1['baro_altitude'].max())
print(file_1['baro_altitude'].min())
```

### Pie Chart of On-Ground Status

```python
# Count the number of occurrences of each value in the onground column
onground_counts = file_1['onground'].value_counts()
# Create a pie chart to visualize the number of flights on the ground vs. in the air
plt.figure(figsize=(6, 6))
plt.pie(onground_counts, labels=onground_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('On-Ground Status')
plt.show()
```

### Create boxplot of ground speed

```python
# Create a boxplot for the ground_speed column
plt.figure(figsize=(8, 6))
plt.boxplot(file_1['ground_speed'].dropna())
plt.title('Boxplot of Ground Speed')
plt.xlabel('Ground Speed')
plt.show()
```

```python
# Print the maximum and minimum values of ground speed for reference
print(file_1['ground_speed'].max())
print(file_1['ground_speed'].min())
```

### Histogram of altitude

```python
# Create a histogram for the altitude column to show distribution
plt.figure(figsize=(10, 6))
plt.hist(file_1['altitude'].dropna(), bins=50, color='blue', alpha=0.7)
plt.title('Histogram of Altitude')
plt.xlabel('Altitude (feet)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Create the same histogram but with a logarithmic scale
plt.figure(figsize=(10, 6))
plt.hist(file_1['altitude'].dropna(), bins=50, color='blue', alpha=0.7)
plt.title('Histogram of Altitude')
plt.xlabel('Altitude (feet)')
plt.ylabel('Frequency')
plt.yscale('log')
plt.show()
```

### Time series of baro altitude

```python
# Plot a graph for 'baro_altitude' over time
file_1['timestamp'] = pd.to_datetime(file_1['timestamp'])
plt.figure(figsize=(12, 6))
plt.plot(file_1['timestamp'], file_1['baro_altitude'])
plt.title('Baro altitude Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Baro altitude (feet)')
plt.show()
```

### Scatterplot of ground speed vs. baro altitude

```python
#Creating a scatter plot of Ground speed vs altitude
plt.figure(figsize=(10, 8))
plt.scatter(file_1['ground_speed'], file_1['baro_altitude'], alpha=0.5)
plt.title('Scatter Plot of Ground Speed vs. Altitude')
plt.xlabel('Ground Speed')
plt.ylabel('Altitude (feet)')
plt.show()
```

### Boxplot of vertical rate

```python
# Boxplot for 'vertical_rate'
plt.figure(figsize=(10, 6))
plt.boxplot(file_1['vertical_rate'].dropna())
plt.title('Boxplot of vertical rate')
plt.xlabel('Vertical rate (meters/second)')
plt.show()
```

```python
# Printing maximum and minimum vertical rates
print(file_1['vertical_rate'].max())
print(file_1['vertical_rate'].min())
```

### Violinplot of vertical rate

```python
# Creating a violin plot for 'vertical_rate' within a specified threshold
threshold = 10000

filtered_values = file_1['vertical_rate'].dropna()
filtered_values = filtered_values[(filtered_values >= -threshold) & (filtered_values <= threshold)]
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(y=filtered_values)
plt.title('Violinplot of vertical rate')
plt.ylabel('Vertical rate (meters/second)')

# Setting y-axis limits to the threshold
plt.ylim(-threshold, threshold)

plt.show()
```
